# **Реляционные БД**



**Реляционные базы данных** — это система организации информации, где данные хранятся в форме таблиц. В этих таблицах каждая строка представляет собой одну запись (например, одного человека), а столбцы — различные характеристики этой записи (например, имя, фамилия, возраст).

Основные особенности реляционных баз данных:

**Таблицы:** Все данные хранятся в структурированных таблицах, что делает информацию легко читаемой и организованной.

**Строки и столбцы:** Таблицы состоят из строк (записей) и столбцов (полей), где каждый столбец имеет определённый тип данных (текст, число, дата и так далее).

**Первичные и внешние ключи:** Эти специальные поля помогают устанавливать связи между таблицами, что позволяет эффективно работать с большими объемами данных.

**SQL:** Это язык запросов, с помощью которого можно легко и быстро извлекать, добавлять, обновлять и удалять данные в базе.

**Нормализация:** Это процесс организации данных для снижения дублирования и улучшения эффективности.

Реляционные базы данных используются повсеместно — от небольших приложений до крупных корпоративных систем, так как они предоставляют удобный и надежный способ хранения и обработки данных.


rbd.jpg






Они называются "реляционными" потому, что основываются на концепции "отношений" (или "связей"). В этом контексте "отношение" — это математический термин, описывающий набор данных, представленных в виде таблицы.

**Таблицы:** В реляционной базе данных информация организована в таблицах (которые также иногда называют "отношениями"). Эти таблицы состоят из строк и столбцов.

**Связи между таблицами:** Реляционные базы данных позволяют создавать связи между различными таблицами на основе общих данных (ключей). Это обеспечивает целостность и консистентность данных в базе.

**Структура данных:** Каждая таблица или "отношение" имеет уникальную структуру, определяемую набором полей (столбцов), каждый из которых имеет определенный тип данных.

Таким образом, термин "реляционный" подчеркивает важность структурированных отношений и связей в организации и управлении данными в этих системах.

### ***Нормализация базы данных***

Это процесс организации структуры данных с целью уменьшить дублирование и зависимость данных. Процесс включает в себя разделение больших таблиц на более мелкие и соединённые между собой таблицы и определение отношений между ними.

Вот несколько ключевых аспектов нормализации:

Уменьшение дублирования данных: Нормализация помогает избежать дублирования данных, что экономит место на диске и упрощает процессы обновления данных.

Повышение целостности данных: Она обеспечивает, что данные будут корректными и надежными путем установления четких правил и структур.

Облегчение обработки данных: С нормализованной структурой данных проще работать, поскольку у вас есть четкое понимание того, где и как хранится каждый фрагмент данных.

Процесс нормализации включает в себя несколько этапов (или "форм"), каждый из которых направлен на решение определенных проблем в структуре данных:

**Первая форма (1NF)**:
Устраняются дубликаты строк, и все элементы в колонке должны быть атомарными (неделимыми).
*  Устраните повторяющиеся группы в отдельных таблицах.
*  Создайте отдельную таблицу для каждого набора связанных данных.
*  Идентифицируйте каждый набор связанных данных с помощью первичного ключа.

**Вторая форма (2NF)**:
Устраняются функциональные зависимости на часть первичного ключа (если первичный ключ составной).
*  Создайте отдельные таблицы для наборов значений, относящихся к нескольким записям.
*  Свяжите эти таблицы с помощью внешнего ключа.

**Третья форма (3NF)**:
Устраняются транзитивные зависимости.
*  Исключите поля, которые не зависят от ключа.



И далее: Существуют и более высокие нормальные формы, например BCNF, 4NF, каждая из которых решает более специфические и редки встречающиеся проблемы.

Нормализация не всегда необходима или полезна. В некоторых случаях, для оптимизации производительности и скорости чтения данных, может потребоваться обратный процесс — денормализация.

## **Типы связей.**

В реляционных базах данных существует несколько основных типов связей:

**Один к одному (1:1):**

Этот тип связи позволяет связать одну запись в первой таблице с одной записью во второй. Пример: каждый человек имеет уникальный паспорт, и каждый паспорт принадлежит одному человеку.

**Один ко многим (1:M):**

В этой связи одна запись в первой таблице может быть связана с несколькими записями во второй таблице. Пример: один автор может написать несколько книг, но каждая книга имеет только одного автора.


**Многие ко многим (M:M):**

Этот тип связи подразумевает, что множество записей в одной таблице может быть связано с множеством записей в другой таблице. Пример: студенты и курсы - каждый студент может посещать несколько курсов, и каждый курс может иметь несколько студентов. Этот тип связи часто реализуется с использованием дополнительной таблицы (таблицы связей).

PS
Стоит еще заметить, что реляционные базы бывают с хранением данных по строкам (PostgreSQL и ко) и по столбцам/колонкам (ClickHouse, Vertica). Колоночные/столбцовые базы лучше подходят для аналитики, в то время как ориентация на строки лучше подходит для транзакционных нагрузок.

# **Документо-ориентированные базы данных.**


Это тип баз данных, в которых информация хранится в виде документов. Под «документом» здесь понимается набор полей и их значений, организованных, например, по принципу «ключ-значение». Документы можно представить как карточки в каталоге, где на каждой карточке записана информация о каком-то объекте, и эта информация может быть разной: текст, числа, списки и так далее.


**Особенности:**
Гибкость структуры: Документы не обязаны иметь одинаковую структуру, то есть каждый документ может содержать уникальный набор полей.
Неструктурированные или полуструктурированные данные: Подходят для хранения данных, структура которых заранее неизвестна или может меняться.
Масштабируемость: Обычно эти базы данных легко масштабируются горизонтально, то есть можно добавлять новые сервера для расширения мощности и объема хранения.
Примеры документо-ориентированных баз данных:


**MongoDB:**

Одна из самых популярных документо-ориентированных баз данных, хранит данные в формате BSON (похож на JSON). Поддерживает запросы, индексацию, агрегацию данных.
CouchDB:

Использует JSON для хранения данных и JavaScript в качестве языка запросов. Ориентирована на удобство использования и масштабируемость.


**RavenDB:**

База данных для .NET платформы, хранит данные в формате JSON. Поддерживает ACID-транзакции, LINQ-запросы.


**Elasticsearch:**

Хотя чаще Elasticsearch используется как поисковый движок, его также можно отнести к документо-ориентированным базам данных, поскольку он хранит и индексирует JSON-документы.


### **Как сделать схему?**

Документо-ориентированные базы данных хранят информацию в виде документов. Наиболее распространенный формат документов — это JSON. Ниже представлен пример схемы для документо-ориентированной базы данных, которая хранит информацию о студентах и их курсах.

```
{
  "_id": "123456",
  "firstName": "Иван",
  "lastName": "Иванов",
  "dateOfBirth": "1995-12-01",
  "address": {
    "street": "Ленина",
    "city": "Москва",
    "postalCode": "123456"
  },
  "courses": [
    {
      "courseName": "Математика",
      "grade": "A",
      "credits": 3
    },
    {
      "courseName": "Физика",
      "grade": "B",
      "credits": 4
    },
    {
      "courseName": "Химия",
      "grade": "C",
      "credits": 2
    }
  ]
}

```

**"_id"**: уникальный идентификатор студента в базе данных.

**"firstName" и "lastName"**: имя и фамилия студента.

**"dateOfBirth"**: дата рождения студента.

**"address"**: объект с детальной адресной информацией студента (улица, город, почтовый индекс).

**"courses"**: массив объектов, каждый из которых содержит информацию о курсах, которые студент проходит или прошел, включая название курса, оценку и количество кредитов.


# **Базы данных ключ-значение**



Это системы, где каждый элемент данных (значение) связан с уникальным ключом. Проще говоря, это как огромный шкаф с ящичками, где на каждом ящичке написан номер (ключ), и внутри каждого ящичка лежит какая-то вещь (значение). Вы можете быстро получить вещь из ящичка, зная его номер.

Примеры баз данных ключ-значение:

**Redis**

Это очень быстрая база данных ключ-значение, которая может хранить строки, списки, множества и другие типы данных. Она часто используется для кеширования данных, создания очередей задач и многого другого.


**Amazon DynamoDB**

Это полностью управляемая база данных ключ-значение, которая обеспечивает быструю и предсказуемую производительность с масштабируемостью. DynamoDB часто используется в облачных приложениях для хранения данных.


**Простой пример использования:**

Вы хотите сохранить номера телефонов своих друзей. В этой базе данных, имя друга будет ключом, а номер телефона — значением. Например:
Ключ: "Анна", Значение: "123-456-7890"
Ключ: "Михаил", Значение: "987-654-3210"

Такие системы хорошо подходят для ситуаций, когда вам нужен быстрый доступ к данным по конкретному ключу, и структура данных довольно простая.







# **Графовые базы данных**


Графовые базы данных — это системы, которые хранят и управляют данными в виде графов. В таких базах данных основной фокус делается на отношениях между данными. В графе есть два ключевых элемента:

**Вершины (узлы):** Это объекты, которые вы хотите представить. Например, люди, места, вещи, события и так далее. Каждый узел содержит информацию об объекте (атрибуты).

**Рёбра (связи):** Это отношения между узлами. Рёбра также могут иметь атрибуты, которые помогают описать характер этой связи.

**Пример:**
Представьте себе социальную сеть. Узлы — это пользователи, а связи — это их дружественные отношения. Таким образом, можно легко отобразить и проанализировать сеть друзей, их взаимосвязи и общие интересы.

Примеры графовых баз данных:

**Neo4j:**

Это одна из самых популярных графовых баз данных. Neo4j предлагает высокую производительность для работы с графами и поддерживает язык запросов Cypher, разработанный специально для работы с графовыми структурами данных.

**ArangoDB:**

Мульти-модельная база данных, которая поддерживает графы, документы и ключ-значение. ArangoDB предлагает гибкий язык запросов и хорошую производительность для работы с большими графами.

**Amazon Neptune:**

Облачная графовая база данных от Amazon, поддерживает работу с огромными графами, хорошо масштабируется и интегрируется с другими сервисами Amazon.
Графовые базы данных отлично подходят для анализа сложных взаимосвязей между объектами, таких как социальные сети, рекомендательные системы, сети связи и другие сферы, где важен анализ взаимосвязей.

### **Как это проецировать (пример)**

**Узлы (вершины):**

Узлы представляют собой пользователей и интересы. Каждый узел-пользователь может иметь атрибуты:

ID (уникальный идентификатор),
Имя,
Возраст и т.д.

**Рёбра (связи):**

Дружба: Связь между двумя пользователями, отображающая их дружественные отношения.
Интерес: Связь между пользователем и интересом, показывающая, что пользователь имеет данный интерес.

Векторные базы данных — это тип систем управления базами данных (СУБД), специализированный на эффективном хранении и поиске векторов. Вектор в данном контексте — это набор чисел, который может представлять различные объекты, например, изображения, тексты, звуки.

**Особенности:**

Быстрый поиск: Векторные базы данных обеспечивают высокоскоростной поиск по схожести (поиск ближайших векторов), что критически важно в задачах машинного обучения и искусственного интеллекта.
Сжатие данных: Они эффективно сжимают векторные данные для экономии места.
Поддержка больших объемов: Позволяют работать с огромными объемами векторных данных.

Примеры векторных баз данных:

**FAISS:**

Разработана Facebook для эффективного поиска по схожести в больших наборах векторов. FAISS активно используется для задач компьютерного зрения и других приложений машинного обучения.


**Annoy (Approximate Nearest Neighbors Oh Yeah):**

Библиотека от Spotify, предназначенная для поиска приближенных ближайших соседей в пространствах большой размерности. Annoy особенно полезен для задач рекомендации.


**Milvus:**

Открытая платформа для поиска и анализа векторов, которая поддерживает различные алгоритмы и способна обрабатывать большие объемы данных. Milvus подходит для широкого спектра приложений, включая поиск по изображениям, рекомендательные системы и биометрическую идентификацию.

Пример использования:

**Шаг 1 - Предобработка данных:**

У вас есть большой корпус текстовых данных (статьи, книги, форумы и т.д.).
Вы обрабатываете каждый документ (или часть документа), используя языковую модель, чтобы преобразовать текст в вектор (эмбеддинг). Этот вектор служит численным представлением смысла или контекста текста.

**Шаг 2 - Индексация векторов:**

Сохраняйте полученные векторы в векторной базе данных (например, FAISS, Annoy или Milvus).
Векторные БД эффективно индексируют эти векторы, позволяя быстро и точно находить наиболее близкие векторы к заданному.

Шаг 3 - Использование в реальном времени:

Пользователь задает вопрос или делает запрос.
Ваша языковая модель преобразует запрос в вектор.
Векторный запрос отправляется в векторную БД, которая находит наиболее близкие векторы (и соответствующие им тексты).
Вы можете возвращать эти тексты пользователю или использовать их как дополнительный контекст для генерации более точного и информативного ответа моделью.
